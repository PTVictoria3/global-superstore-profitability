Ta thấy sales đang là object nhưng thực tế phải là float và order_date,ship_date  phải là datetime thay vì object

```python
clean_df = df.copy()
```

```python
clean_df['sales']=clean_df['sales'].astype(str).str.replace('$','').str.replace(',','')
```

```python
for col in ['sales', 'quantity', 'discount', 'profit', 'shipping_cost']:
    clean_df[col] = pd.to_numeric(clean_df[col], errors='coerce')
```

```python
clean_df['order_date'] = pd.to_datetime(clean_df['order_date'], dayfirst=True)
clean_df['ship_date']  = pd.to_datetime(clean_df['ship_date'],  dayfirst=True)
```

```python
n_cols=clean_df.select_dtypes(include=['number']).columns
print("Các cột có kiểu dữ liệu số:")
for col in n_cols:
    print(f"{col}")
```

```python
o_cols=clean_df.select_dtypes(include=['object']).columns
print("Các cột có kiểu dữ liệu object:")
for col in o_cols:
    print(f"{col}")
    
```

```python
# Loại bỏ khoảng trắng thừa ở đầu và cuối chuỗi trong các cột object
for col in o_cols:
    clean_df[col]=clean_df[col].str.strip()
```

```python
# Kiểm tra và loại bỏ các dòng trùng lặp
print("Các dòng bị trùng lặp : ",clean_df.duplicated().sum())
clean_df=clean_df.drop_duplicates()
```

```python
# Đảm bảo rằng ngày giao hàng không sớm hơn ngày đặt hàng
clean_df[(clean_df['ship_date'] < clean_df['order_date'])]
```

```python
# Tạo cột profit_margin để tính tỷ suất lợi nhuận
clean_df['profit_margin']=np.where(clean_df['sales'] != 0, clean_df['profit'] / clean_df['sales'],0).round(4)
```

```python
# Tạo cột is_loss để đánh dấu các đơn hàng có lợi nhuận âm
clean_df['is_loss']=(clean_df['profit'] < 0).astype(int)
```

```python
def discount_level(discount):
    if discount == 0:
        return 'No'
    elif discount < 0.2:
        return 'Low'
    elif discount < 0.4:
        return 'Medium'
    else:
        return 'High'
```

```python
# Tạo cột discount_level để phân loại mức độ giảm giá ở đây sẽ có 4 mức: No, Low, Medium, High
clean_df['discount_level']=clean_df['discount'].apply(discount_level)
clean_df['discount_level'].value_counts()
```

```python
clean_df['shipping_days']  = (clean_df['ship_date'] - clean_df['order_date']).dt.days
```
